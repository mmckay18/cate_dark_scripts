```python
import os
import numpy as np
from PIL import Image
import glob
import matplotlib.pyplot as plt
import scipy.io
from astropy.table import Table, Column
from astropy.io import fits
import re
```

```python
def natural_sort(l): 
    convert = lambda text: int(text) if text.isdigit() else text.lower() 
    alphanum_key = lambda key: [ convert(c) for c in re.split('([0-9]+)', key) ] 
    return sorted(l, key = alphanum_key)


os.chdir('/grp/hst/wfc3r/mmckay_sci_eclipse/cate17-000b/totality_data/')
total_images=glob.glob('*.tif')
total_images=natural_sort(total_images)
totality_filename = []
totality_mean_values  =[]
totality_med_values   =[]
totality_stddev_values=[]
totality_max_values   =[]
totality_min_values   =[]
file_number=[]
i=0

for image in total_images:
    i+=1
    file_number.append(i)
    im = Image.open(image)
    total_array=np.array(im)
    total_mean = np.mean(total_array)
    total_med = np.median(total_array)
    total_stddev = np.std(total_array)
    total_max = np.max(total_array)
    total_min = np.min(total_array)
    
    totality_filename.append(image)
    totality_mean_values.append(total_mean)  
    totality_med_values.append(total_med)   
    totality_stddev_values.append(total_stddev)
    totality_max_values.append(total_max)   
    totality_min_values.append(total_min)   
    #print(image,total_mean,i)
    
n,bins,patches=plt.hist(totality_mean_values, 50, facecolor='b', alpha=0.50)
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.savefig('mean_histogram.png')
plt.show()





```

```python
import os
import numpy as np
from PIL import Image
import glob
import matplotlib.pyplot as plt
import scipy.io
from astropy.table import Table, Column
import argparse
from astropy.io import fits
from multiprocessing import Pool
```

```python
def natural_sort(l): 
    convert = lambda text: int(text) if text.isdigit() else text.lower() 
    alphanum_key = lambda key: [ convert(c) for c in re.split('([0-9]+)', key) ] 
    return sorted(l, key = alphanum_key)
```

```python
def make_data_txt_file(totality_dark_files):
    totality_darks_filename = []
    totality_darks_mean_values  =[]
    totality_darks_med_values   =[]
    totality_darks_stddev_values=[]
    totality_darks_max_values   =[]
    totality_darks_min_values  =[]
    file_number=[]
    i=0

    totality_dark_files = natural_sort(totality_dark_files)
    for tif in totality_dark_files:
        i+=1
        file_number.append(i)
        image = Image.open(tif)
        total_dark_array=np.array(image)
        total_dark_mean = np.mean(total_dark_array)
        total_dark_med = np.median(total_dark_array)
        total_dark_stddev = np.std(total_dark_array)
        total_dark_max = np.max(total_dark_array)
        total_dark_min = np.min(total_dark_array)
        
        totality_darks_filename.append(tif)
        totality_darks_mean_values.append(total_dark_mean)  
        totality_darks_med_values.append(total_dark_med)   
        totality_darks_stddev_values.append(total_dark_stddev)
        totality_darks_max_values.append(total_dark_max)   
        totality_darks_min_values.append(total_dark_min)   
        #print(image)
    filename=totality_darks_filename[0][0:11]
    #print(filename)
    t=Table()
    t['#File_number'] = file_number
    t['Filename'] = totality_darks_filename
    t['Max Values (Analog-Digital units)'] = totality_darks_max_values
    t['Min Values (Analog-Digital units)'] = totality_darks_min_values
    t['Mean Values (Analog-Digital units)'] = totality_darks_mean_values
    t['Median Values (Analog-Digital units)'] = totality_darks_med_values
    t['STDDEV Values (Analog-Digital units)'] = totality_darks_stddev_values
    t.write('{}_totality_dark_statistics.txt'.format(filename),format='ascii.basic', overwrite=True)
    t.write('{}__totality_dark_statistics_neat.txt'.format(filename),format='ascii.fixed_width', overwrite=True)

```

```python
os.chdir('/grp/hst/wfc3r/mmckay_sci_eclipse/comparing_mmdarks_mjpdarks/cate17-000b/')
Matt_darks = sorted(glob.glob('cate17-*_dark*.fits'))
Myles_darks = sorted(glob.glob('cate17-*_superdark*.fits'))
Myles_darks=natural_sort(Myles_darks)
mean_ratio_list = []
std_ratio_list =[]
median_ratio_list = []
frame = []
a=1
for i,j in zip(Matt_darks, Myles_darks):
    print(i,j)
    Matt_dark_hdu = fits.open(i)
    Matt_dark_data = Matt_dark_hdu[0].data
    Myles_dark_hdu = fits.open(j)
    Myles_dark_data = Myles_dark_hdu[0].data
    mean_ratio = np.mean(Matt_dark_data) / np.mean(Myles_dark_data)
    std_ratio = np.std(Matt_dark_data) / np.std(Myles_dark_data)
    median_ratio = np.median(Matt_dark_data) / np.median(Myles_dark_data)
    frame.append(a)
    #print (np.mean(Matt_dark_data),np.mean(Myles_dark_data))
    
    print('frame: ',a, 'Mean Ratio: ', mean_ratio,'STDDEV Ratio: ', std_ratio, 'Median Ratio: ', median_ratio)
    a+=1
    mean_ratio_list.append(mean_ratio)
    std_ratio_list.append(std_ratio)
    median_ratio_list.append(median_ratio)

plt.scatter(frame,mean_ratio_list,label='Mean')
plt.plot(frame,mean_ratio_list)
plt.scatter(frame,std_ratio_list,label='STDDEV')
plt.plot(frame,std_ratio_list)
plt.scatter(frame,median_ratio_list,label='Median')
plt.plot(frame,median_ratio_list)
plt.title('{} MJP darks mean / MM darks mean'.format(Myles_darks[0][:11]))
plt.ylabel('Ratio')
plt.xlabel('Dark Frame')
plt.legend()
#plt.axis([0,9,0.9995,1.0005])
plt.savefig('{} ratio plot.png'.format(Myles_darks[0][:11]))
plt.show()
plt.clf()
```
